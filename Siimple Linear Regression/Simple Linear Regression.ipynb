## Simple Linear Regression - Salary Prediction

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Salary_Data.csv')
x = dataset.iloc[:, 0:1].values #YearsExperience
y = dataset.iloc[:, 1].values   # Salary
```

__train_test_split(*arrays, test_size=None, train_size=None, random_state=None, shuffle=True, stratify=None)__: Split arrays or matrices into random train and test subsets

```python
# Splitting the dataset into tarining ans testing dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.25, random_state=42)
```

__LinearRegression(*, fit_intercept=True, normalize=False, copy_X=True, n_jobs=None, positive=False,)__: Ordinary least squares Linear Regression. <br>
<br>
LinearRegression fits a linear model with coefficients w = (w1, ..., wp)
to minimize the residual sum of squares between the observed targets in
the dataset, and the targets predicted by the linear approximation.

```python
# Creating the Linear Regression model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

1. __regressor.fit(X, y, sample_weight=None)__: Fit linear model
2. __regressor.predict(X)__: Predict using the linear model.

```python
# Triaing the model with the training dataset
# and then predicting y value from the test dataset
regressor.fit(x_train, y_train) # train model
y_pred = regressor.predict(x_test) # predict y
```

__metrics.mean_squared_error(y_true, y_pred, *, sample_weight=None, multioutput='uniform_average', squared=True,)__: Mean squared error regression loss

```python
# Computing the mean squared error between the actual and predicted value of y
from sklearn import metrics
print("Error=", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# Predict salary for 2.3 year experience
print("Salary : ", regressor.predict([[2.3]]))
```

```python
# Visualizing the actual training data
plt.plot(x_train, y_train, 'r*')
#plt.plot(x_train, y_train, 'g')
```

```python
# Predicting y value for training dataset
y_pred_train = regressor.predict(x_train)
```

```python
# Visualizing the actual and predicted data
plt.plot(x_train, y_pred_train)
# plt.plot([[2.3]], regressor.predict(np.array([[2.3]])), 'g^')
plt.title("Employee Data")
plt.xlabel("Years Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Training Set
plt.scatter(x_train, y_train, color='red')
plt.plot(x_train, regressor.predict(x_train), color='blue')
plt.title('Salary vs Experience(Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Testing Set
plt.scatter(x_test, y_test, color='red')
plt.plot(x_train, regressor.predict(x_train), color='blue')
plt.title('Salary vs Experienc( Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Calculating the coefficient and intercept of the line
print("Intercept: ", regressor.intercept_)
print("Co-efficient:", regressor.coef_)
```
