## K-Nearest Neighbours Classification

```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
dataset = pd.read_csv('Social_Network_Ads.csv')
dataset.head()
```

```python
X = dataset.iloc[:, [2,3]].values # Age and EstimatedSalary
y = dataset.iloc[:, 4].values  # Purchased
```

```python
# Splitting the dataset into Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
df_xtrain = pd.DataFrame(X_train)
df_xtest = pd.DataFrame(X_test)
df_ytrain = pd.DataFrame(y_train)
df_ytest = pd.DataFrame(y_test)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# Fitting K-NN to train the Training set
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(weights='distance')
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
print("My Confusion Matrix:", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy: ", accuracy_score(y_test, y_pred))
```

```python
# new customer prediction age=28 salary=119000
k = classifier.predict(sc.transform([[28, 119000]]))
print("New customer class: ", k)
if k[0] == 0:
    print("Not purchase type")
else:
    print("Purchase type")
```

```python
# plotting the graph
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
arranged_ages = np.arange(start = X_set[:, 0].min(), stop = X_set[:, 0].max(), step=0.01)
arranged_salaries = np.arange(start = X_set[:, 1].min(), stop = X_set[:, 1].max(), step=0.01)
X1, X2 = np.meshgrid(arranged_ages, arranged_salaries)

plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape), alpha = 0.8, cmap = ListedColormap(('orange', 'blue')))
```

```python
# Plotting Test set:
plt.xlim(X1.min(), X2.max())
plt.ylim(X2.min(), X2.max())
plt.plot(sc.transform([[28,119000]])[:,0],
        sc.transform([[28,119000]])[:,1],'g*')
plt.title('Customer')
plt.xlabel('Age')
plt.ylabel('Salary')
# plt.legend()
plt.show()
```
