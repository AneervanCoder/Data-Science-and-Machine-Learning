## Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

1. __plt.plot(*args, scalex=True, scaley=True, data=None, **kwargs)__: Plot y versus x as lines and/or markers.
2. __plt.title(label, fontdict=None, loc=None, pad=None, *, y=None, **kwargs)__: Set a title for the axes.
3. __plt.xlabel(xlabel, fontdict=None, labelpad=None, *, loc=None, **kwargs)__: Set the label for the x-axis.
4. __plt.ylabel(ylabel, fontdict=None, labelpad=None, *, loc=None, **kwargs)__: Set the label for the y-axis.
5. __plt.show(close=None, block=None)__: Display all open figures.
6. __plt.legend(*args, **kwargs)__: *args, **kwargs

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y,'^')
plt.title("x-y values")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.show()
```

```python
# if we don't specify the marker by default it is a line graph
plt.plot(x,y, color = 'green')
plt.title("x-y values")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
overs = [1,2,3,4,5]
team_a = [11,12,23,14,34]
team_b = [10,13,12,20,12]

# plotting the graph of Team A only
plt.plot(overs, team_a)
plt.title("Team Performance")
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.show()
```

```python
# Plotting the graph of Team A and Team B both
plt.plot(overs, team_a, label = "Team A")
plt.plot(overs, team_b, label = "Team B")
plt.title("Team Performance")
plt.xlabel("overs")
plt.ylabel("runs")
plt.legend()
plt.show()
```

```python
goal_types = 'Penalties', 'Field Goals', 'Free Kicks'
goals = [12,38,7]
mycolors = ['y','r','b']
```

```python
plt.pie(goals, labels=goal_types, colors=mycolors, shadow=True, explode=(0.0,0.0,0.2), autopct = "%.2f%%")
plt.show()
```

```python
data = pd.read_csv("player_data.csv")
g_type = list(data.columns)

# start row  index : end row index
# start column index :  end column index
n_goals = list(data.iloc[1,:])
mycolors = ['y','r','b']
```

```python
plt.pie(n_goals, labels=g_type, colors=mycolors, shadow=False,
       explode=(0.0,0.0,0.0), autopct='%.2f%%')
plt.show()
```

```python
n_goals1=list(data.iloc[0,:])
plt.pie(n_goals1, labels=g_type, colors=mycolors, shadow=False,
       explode=(0.0,0.0,0.0), autopct='%.2f%%')
plt.show()
```
